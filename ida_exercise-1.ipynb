```r
# Exercise-1
data <- c(1, 2, 3, 3, 2, 45, 45, 56, 67, 55, 67, 56, 67, 68, 55, 54, 43, 32, 22, 33)

# (a) Calculate the Arithmetic Mean
mean_value <- mean(data)
cat("Arithmetic Mean:", mean_value, "\n")

# (b) Calculate the Median
median_value <- median(data)
cat("Median:", median_value, "\n")

# (c) Calculate the Mode
# Custom function to calculate mode
get_mode <- function(x) {
  uniq_x <- unique(x)
  uniq_x[which.max(tabulate(match(x, uniq_x)))] # Returns the mode value
}

mode_value <- get_mode(data)
cat("Mode:", mode_value, "\n")

```
